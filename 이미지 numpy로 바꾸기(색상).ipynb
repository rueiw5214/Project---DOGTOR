```python
import matplotlib.pyplot as plt #그래프 출력용
from sklearn.tree import DecisionTreeClassifier #의사 결정 나무 분류기
import numpy as np #넘파이
import pandas as pd #판다스
from sklearn.ensemble import RandomForestClassifier #랜덤 포레스트 분류기
from sklearn.naive_bayes import GaussianNB #가우시안 나이브 베이즈
from sklearn.neighbors import KNeighborsClassifier #K-neighbors 분류기
from sklearn.ensemble import AdaBoostClassifier, VotingClassifier, BaggingClassifier
#Adaboost 분류기,Voting 분류기, Bagging 분류기 
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis #???
from sklearn.svm import SVC #Support Vector Machine 분류기
from xgboost import XGBClassifier #XGBoost 분류기
import os #경로를 지정하는데 사용하는 라이브러리
from PIL import Image #이미지 작업용 라이브러리
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_confusion_matrix
#사이킷런 기반 머신러닝 모델들의 평가 지표(정확도, 혼동 행렬,분류 리포트)
import plotly.express as px #시각화 전용 라이브러리
import cv2 #이미지 작업을 위한 openCV
```

```python
img_path = r"C:\Users\user\imagedata\color_cluster_sample" #이미지 경로 지정

folder_list = ['red', 'black', 'yellow', 'green', 'gray'] #폴더 리스트(폴더 이름)
img_list_np = [] #배열화 된 이미지를 저장하기 위한 리스트

labels_str = [] #색상을 추가할 라벨링용 리스트
labels = [] #번호를 추가할 라벨링용 리스트
for color in folder_list:
    img_list = os.listdir(img_path +'/'+ color) #이미지 경로의 color 폴더안 파일의 리스트
    img_list_png = [img for img in img_list if img.endswith(".png")] #위 리스트의 .png 확장자 파일 리스트
    for i in img_list_png: 
        img = cv2.imread(img_path +'/'+ color + '/' + i) #경로상의 이미지 읽어오기
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #이미지 색상을 BGR에서 RGB로 변경
#         img = Image.open(img_path +'/'+ color + '/' + i).convert('RGB')
        # img = img.resize((256,256)) #이미지 크기 조절
        img_array = np.array(img) #이미지를 numpy 배열화.
        img_list_np.append(img_array) #배열화 된 이미지를 리스트에 저장
        print(i, " 추가 완료 - 구조: ", img_array.shape) #리스트에 추가 되었다고 알림
        print(img_array.T.shape) #이미지 배열의 전치된 배열의 크기 출력
        labels_str.append(color) # 라벨링용 리스트에 색상 추가

for i in labels_str:
    temp = i.replace('red', '0').replace('black', '1').replace('yellow', '2').replace('green', '3').replace('gray', '4')
    #라벨링용 리스트의 색에 따라 문구를 숫자로 변경
    labels.append(temp)
    #번호 라벨링용 리스트에 저장

img_np = np.array(img_list_np) #이미지 배열화된 이미지 배열
np.save('X_train_color_sample', img_np) #X_train_color_sample 파일에 img_np 저장
labels = [int(x) for x in labels] # 숫자로 해야 학습 가능 숫자로 저장
label_np = np.array(labels) # 리스트를 배열화
np.save('y_train_color_sample', label_np) #y_train_color_sample 파일에 label_np 저장
print(img_np.shape) # 구조 확인해보장
print("결과: 저장완료") #완료되면 문장 출력
```
